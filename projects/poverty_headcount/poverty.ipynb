```python
import pandas as pd

poverty = pd.read_csv('API_SI.POV.DDAY_DS2_en_csv_v2_385351.csv')
poverty

population = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_382278.csv')
population
```

```python
years = [str(i) for i in range(1960, 2019)]

# most recent values for poverty line
mrvl_pov = []
# most recent values for population
mrvl_pop = []
# most recent year for poverty line
mryl_pov = []
# most recent year for population
mryl_pop = []

for i in range(len(poverty)):
    pov_row = poverty[years][poverty.index == i]
    pov_row = pov_row.dropna(axis=1, how='all')
    
    pop_row = population[years][population.index == i]
    pop_row = pop_row.dropna(axis=1, how='all')
    
    if not pov_row.empty:
        mrd_pov = pov_row[max(pov_row.columns)].iloc[0]  
        mry_pov = max(pov_row.columns)
    else:
        mrd_pov = None
        mry_pov = ""
        
    if not pop_row.empty:
        mrd_pop = pop_row[max(pop_row.columns)].iloc[0]
        mry_pop = max(pop_row.columns)
    else:
        mrd_pop = None
        mry_pop = ""
        
    mrvl_pov.append(mrd_pov)
    mrvl_pop.append(mrd_pop)
    mryl_pov.append(mry_pov)
    mryl_pop.append(mry_pop)


poverty['most_recent_year_pov'] = mryl_pov
poverty['most_recent_value_pov'] = mrvl_pov
poverty['most_recent_year_pop'] = mryl_pop
poverty['most_recent_value_pop'] = mrvl_pop
```

```python
poverty[['country_name', 'most_recent_year_pov', 'most_recent_value_pov', 'most_recent_year_pop', 'most_recent_value_pop']].to_csv('poverty_headc.csv', index=False)
```
