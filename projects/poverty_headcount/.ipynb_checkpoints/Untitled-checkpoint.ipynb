```python
import pandas as pd

df = pd.read_csv('API_SI.POV.DDAY_DS2_en_csv_v2_385351.csv')
df
```

```python
#a = df[min(i for i )]
# yl = [i for i in df.columns if int
# min(df[i] for i df.columns)
y = [str(i) for i in range(1960, 2019)]
l = []
yy = []
for i in range(len(df)):
    b = df[y][df.index == i]

    b = b.dropna(axis=1, how='all')
    if not b.empty:
        
        k = b[max(b.columns)].iloc[0]  
        y0 = max(b.columns)
    else:
        k = None
        y0 = ""
        
    l.append(k)
    yy.append(y0)


df['most_recent_year'] = yy
df['most_recent_value'] = l
```

```python
df[['Country Name', 'most_recent_year', 'most_recent_value']].to_csv('poverty_headc.csv', index=False)
```
